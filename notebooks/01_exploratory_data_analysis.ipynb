# 01 - Exploratory Data Analysis

## Objective

This notebook explores the credit risk dataset used for Probability of Default modeling.

The goal is to understand the structure of the data, analyze missing values, study the target distribution and identify important patterns before modeling.

## Confidentiality note

The original dataset comes from a professional banking environment and is not included in this repository.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_PATH = "../data/echantillon_etu_tr_2.csv"

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## Data cleaning preview

The ID column is removed because it is only a client identifier and does not provide predictive information.

```python
df_clean = df.copy()

if "ID" in df_clean.columns:
    df_clean = df_clean.drop(columns=["ID"])

df_clean.head()
```

```python
df_clean.isnull().sum()
```

```python
if "ancienneté1" in df_clean.columns:
    df_clean["ancienneté1"] = df_clean["ancienneté1"].fillna(df_clean["ancienneté1"].mean())

if "engagement2" in df_clean.columns:
    df_clean["engagement2"] = df_clean["engagement2"].fillna(df_clean["engagement2"].mean())

if "type_imp" in df_clean.columns:
    df_clean["type_imp"] = df_clean["type_imp"].fillna("No incident")

df_clean.isnull().sum()
```

## Target distribution

The target variable represents the default indicator. In credit risk, the default class is usually less frequent than the non-default class.

```python
target_column = "défaut"

df_clean[target_column].value_counts()
```

```python
df_clean[target_column].value_counts(normalize=True)
```

```python
df_clean[target_column].value_counts().plot(kind="bar", figsize=(6, 4))
plt.title("Target Distribution")
plt.xlabel("Default")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

## Numerical variable distributions

Financial variables often show asymmetry, outliers and heterogeneous scales.

```python
df_clean.hist(figsize=(12, 10), bins=15)
plt.tight_layout()
plt.show()
```

## Categorical variable analysis

```python
if "type_imp" in df_clean.columns:
    df_clean["type_imp"].value_counts().plot(kind="bar", figsize=(8, 4))
    plt.title("Distribution of Payment Incident Types")
    plt.xlabel("Type of incident")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
if "type_imp" in df_clean.columns:
    pd.crosstab(df_clean["type_imp"], df_clean["défaut"], normalize="index") * 100
```

## Boxplot example

This plot compares the distribution of financial commitment values by default status.

```python
if "engagement1" in df_clean.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x="défaut", y="engagement1", data=df_clean)
    plt.title("Engagement 1 by Default Status")
    plt.grid(True)
    plt.show()
```

## Correlation analysis

The variable `tx_imp` is removed if it has very low variance or limited information.

```python
df_corr = df_clean.copy()

if "tx_imp" in df_corr.columns:
    df_corr = df_corr.drop(columns=["tx_imp"])

columns_to_drop = [col for col in ["défaut", "type_imp", "Segment"] if col in df_corr.columns]
df_numeric = df_corr.drop(columns=columns_to_drop)

plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## EDA conclusion

The exploratory analysis highlights:

- A class imbalance between default and non-default clients
- Missing values in some variables
- Strong heterogeneity in financial variables
- Potential outliers in financial commitments
- Low correlations between most explanatory variables
- The need for preprocessing before modeling
